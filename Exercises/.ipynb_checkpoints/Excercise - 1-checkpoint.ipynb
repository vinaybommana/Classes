# Using math module in python

  Write a program that:
  - Stores two numbers in variables a and b
  - Uses math.sqrt() to print the square root of each
  - Uses math.pow() to print a raised to the power of b
  - Uses math.gcd() to print their greatest common divisor
```
  Example output (a = 36, b = 12):
  Square root of 36: 6.0
  Square root of 12: 3.4641016151377544
  36 raised to 12: 4738381338321616896.0
  GCD of 36 and 12: 12
```
