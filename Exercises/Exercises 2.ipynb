## Greeting by Time of Day
     Given an hour variable (0-23), print a greeting using an f-string.
     - Use "Good morning" for 5-11,
     - "Good afternoon" for 12-17,
     - "Good evening" for 18-21, and
     - "Good night" otherwise.
     - Format: "Good morning, it's 8 o'clock!"

## Grade Calculator 

Given a score (0-100), determine the letter grade (A: 90+, B: 80-89, C: 70-79, D: 60-69, F: below 60). 
Print: "You scored 85. Your grade is B."

## Even/Odd Classifier 

Given a number n, use % to check if it's even or odd. Print "7 is odd" or "4 is even". 
Bonus: also include whether it's positive, negative, or zero.

## Password Strength

Given a password string, check its length with len(). Under 6 chars is "Weak", 6-11 is "Medium", 12+ is
  "Strong". Print: "'hello' is too short. Weak password."
